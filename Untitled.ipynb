```python
data = []

movie_id = None
count = 0

with open(file_path, "r") as f:

    for line in tqdm(f):

        line = line.strip()

        if ":" in line:
            movie_id = int(line.replace(":", ""))

        else:

            user_id, rating, date = line.split(",")

            data.append([int(user_id), movie_id, int(rating), date])

            count += 1

            if count >= rows_limit:
                break
```

# Recommendation Model Development

We implement and compare two recommendation approaches:

1. User-Based Collaborative Filtering
2. Matrix Factorization using SVD

The models are evaluated using RMSE and recommendation quality.

```python
from surprise import Dataset
from surprise import Reader

from surprise import KNNBasic
from surprise import SVD

from surprise.model_selection import train_test_split

from surprise import accuracy
```

```python
reader = Reader(rating_scale=(1, 5))

data = Dataset.load_from_df(df[["User_ID", "Movie_ID", "Rating"]], reader)
```

```python
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)
```

```python
print("Train Ratings:", trainset.n_ratings)
print("Test Ratings:", len(testset))
```

## Model 1: User-Based Collaborative Filtering

```python
sim_options = {"name": "cosine", "user_based": False}
```

```python
item_cf = KNNBasic(sim_options=sim_options)

item_cf.fit(trainset)
```

```python
predictions_item = item_cf.test(testset)
```

```python
rmse_item = accuracy.rmse(predictions_item)

print("Item-Based CF RMSE:", rmse_item)
```

## Model 2: SVD Matrix Factorization

```python
svd_model = SVD(random_state=42)
```

```python
svd_model.fit(trainset)
```

```python
predictions_svd = svd_model.test(testset)
```

```python
rmse_svd = accuracy.rmse(predictions_svd)

print("SVD RMSE:", rmse_svd)
```

```python
comparison = pd.DataFrame(
    {"Model": ["Item-Based CF", "SVD"], "RMSE": [rmse_item, rmse_svd]}
)

comparison
```

```python
plt.figure(figsize=(6, 4))

sns.barplot(data=comparison, x="Model", y="RMSE", palette="Set2")

plt.title("Model Comparison (RMSE)")
plt.show()
```

### Recommendation Analysis

The SVD model was used to generate personalized recommendations.

Movies not previously rated by the user were identified and assigned predicted ratings.

The top-ranked movies were recommended as potential viewing choices.

The recommendations demonstrate how collaborative filtering can uncover user preferences and suggest relevant content even when explicit ratings are unavailable.

```python
sample_users = np.random.choice(df["User_ID"].unique(), size=500, replace=False)

print("Sample Users:", len(sample_users))
```

```python
relevant_movies = {}

for user in sample_users:

    relevant = set(df[(df["User_ID"] == user) & (df["Rating"] >= 4)]["Movie_ID"])

    relevant_movies[user] = relevant
```

```python
def average_precision_at_k(recommended, relevant, k=10):

    recommended = recommended[:k]

    score = 0.0
    hits = 0

    for i, movie in enumerate(recommended):

        if movie in relevant:

            hits += 1
            score += hits / (i + 1)

    if len(relevant) == 0:
        return 0

    return score / min(len(relevant), k)
```

```python
map_scores = []

all_movies = df["Movie_ID"].unique()

for user in sample_users:

    rated_movies = set(df[df["User_ID"] == user]["Movie_ID"])

    unrated_movies = np.setdiff1d(all_movies, list(rated_movies))

    preds = []

    for movie in unrated_movies:

        pred = svd_model.predict(user, movie)

        preds.append((movie, pred.est))

    preds = sorted(preds, key=lambda x: x[1], reverse=True)

    top10 = [movie for movie, score in preds[:10]]

    ap = average_precision_at_k(top10, relevant_movies[user], k=10)

    map_scores.append(ap)
```

```python
map10 = np.mean(map_scores)

print("MAP@10:", round(map10, 4))
```

# Similar Movie Recommendation

```python
pivot = df.pivot_table(index="User_ID", columns="Movie_ID", values="Rating")

pivot.head()
```

```python
movie_counts = df.groupby("Movie_ID").size().sort_values(ascending=False)

movie_counts.head()
```

```python
target_movie = movie_counts.index[0]

print(target_movie)
```

```python
target_ratings = pivot[target_movie]

similar_movies = pivot.corrwith(target_ratings)

similar_movies = pd.DataFrame(similar_movies, columns=["Correlation"])

similar_movies.dropna(inplace=True)

similar_movies = similar_movies.sort_values("Correlation", ascending=False)

similar_movies.head(10)
```

```python
similar_movies.reset_index(inplace=True)

similar_movies.columns = ["Movie_ID", "Correlation"]

similar_movies = similar_movies.merge(movies[["Movie_ID", "Title"]], on="Movie_ID")

similar_movies.head(10)
```
